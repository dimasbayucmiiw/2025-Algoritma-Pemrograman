Fungsi dengan *args dan **kwargs

```python
def jumlahkan(*args):
  return sum(args)
def biodata(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")
print("Total:", jumlahkan(1,2,3,4,5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

Fungsi Decorator

```python
def decorator_sapa(fungsi):
  def wrapper():
    print("Sebelum fungsi dijalankan...")
    fungsi()
    print(f"Setelah fungsi dijalankan...")
  return wrapper
@decorator_sapa
def sapa():
  print("Halo, Mahasiswa Polines!")
sapa()
```

Fungsi Decorator

```python
def logger(func):
    def wrapper():
        print("Memulai fungsi...")
        func()
        print("Fungsi selesai dijalankan.")
    return wrapper
@logger
def halo():
    print("Halo semuanya!")
halo()
```

Studi Kasus Nyata-Program Nilai

```python
def input_nilai():
    return [float(x) for x in input("Masukkan nilai mahasiswa (pisahkan dengan spasi): ").split()]
def hitung_rata(data):
    return sum(data) / len(data)
def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"
nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)
nilai_mahasiswa = [80, 70, 90]
rata = hitung_rata(nilai_mahasiswa)
print("Nilai rata-rata:", rata)
print("Status kelulusan:", status_kelulusan(rata))
print("Nilai akhir (bobot 40% tugas, 60% ujian):", nilai_akhir(75, 85))
```
