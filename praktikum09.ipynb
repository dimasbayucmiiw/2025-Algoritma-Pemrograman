Model Buatan Sendiri

```python
%%writefile konversi.py
def celcius_ke_fahrenheit(c):
    return (c * 9/5) + 32

def fahrenheit_ke_celcius(f):
    return (f - 32) * 5/9
```

```python
import konversi
c = 25
f = konversi.celcius_ke_fahrenheit(c)
print(f"{c}°C = {f}°F")
f2 = 77
c2 = konversi.fahrenheit_ke_celcius(f2)
print(f"{f2}°F = {c2:.2f}°C")
```

Studi Kasus Mini

```python
%%writefile laporan.py
def tampilkan_ringkasan(total_soal, skor):
    print("\n===== RINGKASAN KUIS =====")
    print(f"Jumlah soal : {total_soal}")
    print(f"Skor akhir  : {skor}/{total_soal}")
    print("===========================")
```

```python
import random
import laporan
def kuis_penjumlahan(jumlah_soal=5):
    skor = 0
    for i in range(1, jumlah_soal + 1):
        a = random.randint(1, 10)
        b = random.randint(1, 10)
        print(f"Soal {i}: {a} + {b} = ?")
        jawaban = int(input("Jawab: "))
        if jawaban == a + b:
            print("Benar!\n")
            skor += 1
        else:
            print(f"Salah! Jawaban yang benar adalah {a + b}\n")
    laporan.tampilkan_ringkasan(jumlah_soal, skor)
kuis_penjumlahan()
```
